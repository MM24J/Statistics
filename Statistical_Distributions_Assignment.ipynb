```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mean = 0
```

```python
sd = 1
n = 1000
```

```python
plt.figure(figsize=(10,5))
```

```python
df = pd.DataFrame(np.random.normal(mean, sd, n))
```

```python
df.hist()
```

```python
# It appears that the data follows the normal distribution as it is symmetric around the mean.

print("Mean of data: {}".format(df.mean().values))
print("Points located 2 SD's from the mean: {} {}".format(2*df.std().values,-2*df.std().values))
```

```python
# 95% of the data should fall within 2 SD's from the mean

print("95% of the data corresponds to {} {}".format(np.percentile(df,2.5), np.percentile(df, 97.5)))
```

```python
# Since the percentage is rounded, we can say that these values are close enough to consider 95% of the data to be within 2 SD's of the mean.
```

```python
data_sample = np.random.randint(10,40,100)
```

```python
normal_data = np.random.normal(np.mean(data_sample),
                              np.std(data_sample), len(data_sample))
```

```python
data_sample.sort()
```

```python
normal_data.sort()
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.scatter(normal_data, data_sample)
plt.plot([np.min(normal_data), np.max(normal_data)],
    [np.min(data_sample), np.max(data_sample)], c='red')
plt.xlabel("Sample Data")
plt.ylabel("Normal Data")
```

```python
# It appears that the scatterplot of the blue data deviates too much from the 45 degree red line to be considered normal data.
```
